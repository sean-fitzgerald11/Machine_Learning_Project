```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
# this shell for this code was provided by the AirNow.Org website on the documentation page for this specific
# API endpoint (https://docs.airnowapi.org/Data/docs)

# API parameters
options = {}
options["url"] = "https://airnowapi.org/aq/data/"
options["start_date"] = ''
options["start_hour_utc"] = "00"
options["end_date"] = ''
options["end_hour_utc"] = "23"
options["parameters"] = "o3,pm25,pm10,co,no2,so2"
options["bbox"] = "-109.060253,36.992424,-102.041524,41.003444"
options["data_type"] = "B"
options["format"] = "text/csv"
options["api_key"] = "60B6D35F-3CA9-4BAA-BCAA-FCBB164E113D"
options['verbose'] = '1'

# API request URL
def generate_URL(options):
    request_url = options["url"] \
    + "?startdate=" + options["start_date"] \
    + "t" + options["start_hour_utc"] \
    + "&enddate=" + options["end_date"] \
    + "t" + options["end_hour_utc"] \
    + "&parameters=" + options["parameters"] \
    + "&bbox=" + options["bbox"] \
    + "&datatype=" + options["data_type"] \
    + "&format=" + options["format"] \
    + "&api_key=" + options["api_key"]\
    + "&verbose=" + options["verbose"]
    return request_url
```

```python
start_date = datetime.strptime('2024-01-01', '%Y-%m-%d')
end_date = datetime.strptime('2024-12-31', '%Y-%m-%d')
delta = timedelta(days=3)
i = 72

while start_date <= end_date:
    chunk_start = start_date
    chunk_end = min(start_date + timedelta(days=2), end_date)
    
    start_str = chunk_start.strftime('%Y-%m-%d')
    end_str = chunk_end.strftime('%Y-%m-%d')

    options['start_date']= start_str
    options["end_date"] = end_str

    url=generate_URL(options)
    name = f'airnow_data_{i}.csv'
    
    try:
        response = requests.get(url)
        response.raise_for_status()
        with open(name, "w", encoding="utf-8") as f:
            f.write(response.text)
        print(f'Chunk {i} written to {name}')
              
    except requests.exceptions.HTTPError as e:
        print(f'Error on chunk {i}; Status Code: {e.response.status_code}; Reason: {e.response.reason}')

    start_date += delta
    i += 1
```
